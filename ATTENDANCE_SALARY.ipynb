## Format the file

```python
df1=df1.fillna(0)
```

```python
df1=df1.rename(columns = lambda x: 'A'+ str(x))
```

```python
df1=df1.drop(index=df1[df1['A0']=='Company:'].index)
```

```python
list(set(list(df1["A0"].values)))
```

```python
rows_to_delete=[0,'Department:','Status','Duration','OT','Late By','Shift','Early By','Days']

for i in rows_to_delete:
    df1=df1.drop(index=df1[df1['A0']==i].index)
```

```python
df1
```

```python
def split_empcode(x):
    try:
        new=x.split(":")[0]
        return(new)
    except:
        return(x)
    
def split_empname(x):
    try:
        new=x.split(":")[1]
        return(new)
    except:
        return(x)

df1["Employee Code"]=df1["A2"].apply(lambda x:split_empcode(x))

df1["Employee Name"]=df1["A2"].apply(lambda x:split_empname(x))
```
